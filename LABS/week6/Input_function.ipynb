```python
def play_song(track_id):
    return IFrame(src="https://open.spotify.com/embed/track/"+track_id,
       width="320",
       height="80",
       frameborder="0",
       allowtransparency="true",
       allow="encrypted-media",
      )
```

```python
def get_playlist_tracks(spotify, playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id,market="DE")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
def get_audio_features(playlist_id):
    list_of_audio_features=[]
    for i in range(0, len(tracks)):
        list_of_audio_features.append(sp.audio_features(get_playlist_tracks[i]["track"]["id"])[0])
    return list_of_audio_features
```

```python
import pickle

with open ("kmeans5.pickle", "rb") as f:
    kmeans5 = pickle.load(f)
    
with open ("scaler.pickle", "rb") as f:
    scaler = pickle.load(f)
```

```python
def recommender(song):
    if song in list(billboard['title']) or fuzz.partial_ratio(song, list(billboard['title']))>90:
        top_song=random.choice(billboard['title'])
        result_top = sp.search(q=top_song, limit=1, market='DE')
        track_id_top = result_top['tracks']['items'][0]['id']
        return play_song(track_id_top)
    else:
        result = sp.search(q=song, limit=1, market='DE')
        track_id = result["tracks"]["items"][0]["id"]
        song_features=sp.audio_features(track_id)
        df_features = pd.DataFrame(song_features)
        df_features_n = df_features.select_dtypes(include=np.number)
        features_scaled = scaler.transform(df_features_n)
        df_features_scaled = pd.DataFrame(features_scaled, columns=df_features_n.columns)
        cluster = kmeans5.predict(df_features_scaled)[0]
        rec_song = list(np.where(playlist["cluster"]==cluster)[0])
        recommendation = random.choice(rec_song)
        id_recommendation = playlist.iloc[recommendation, 0]
        print (song)
        return play_song(id_recommendation)
        
```
