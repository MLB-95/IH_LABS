```python
def get_playlist_tracks(spotify, playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id,market="DE")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
def extract_features(ids):
    features_by_id = []
    i = 0
    while i < len(ids):
        features_by_id.append(sp.audio_features(ids[i])[0])
        i += 1
    return features_by_id
```

```python
def new_playlist(playlist_id):
    playlist=sp.user_playlist_tracks("Spotify", playlist_id,market="DE")
    playlist_tracks=get_playlist_tracks("Spotify", playlist_id)
    number_songs=len(playlist_tracks)
    list_of_audio_features=[]
    for item in range(0,int(number_songs)):
        list_of_audio_features.append(sp.audio_features(playlist_tracks[item]["track"]["id"])[0])
    df_playlist = pd.DataFrame(list_of_audio_features)
    return df_playlist
```
