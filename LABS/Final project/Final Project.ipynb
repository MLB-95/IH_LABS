```python
def categorical_city(data_clean):
    city_categories = {
        'Rome': 0,
        'Berlin': 1,
        'London': 2,
        'Lisbon': 3,
        'Amsterdam': 4,
        'Barcelona': 5,
        'Budapest': 6,
        'Athens': 7,
        'Paris': 8,
        'Vienna': 9
    }
    
    categories = []
    for city in data_clean['city']:
        if city in city_categories:
            categories.append(city_categories[city])
        else:
            categories.append(-1)  # or some other value to represent unknown categories
    
    return categories
```

```python
def categorical_room(data_clean):
    room_type_cat = {
        'Private room' : 0,
        'Entire home/apt' :1,
        'Shared room' : 2
    }
    
    room_type_categories = []
    for room_type in data_clean['room_type']:
        if room_type in room_type_cat:
            room_type_categories.append(room_type_cat[room_type])
        else:
            room_type_categories.append(-1)
    return room_type_categories
```

```python
data_clean['city_categories'] = categorical_city(data_clean)
```

```python
data_clean['room_categories'] = categorical_room(data_clean)
```
